## Using tweepy to access Twitter data
- Tweepy supports accessing Twitter via Basic Authentication and the newer method, OAuth. Here I set up an access for Twitter API using tweepy with OAuth to collect tweet data.

```python
# Consumer keys and access tokens, used for OAuth
consumer_key = ''
consumer_secret = ''
access_token = ''
access_secret = ''

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth_handler=auth, 
                 wait_on_rate_limit=True, 
                 wait_on_rate_limit_notify=True)
```

```python
# counts of tweet ids in df dataframe
tweet_ids = df.tweet_id.values
len(tweet_ids)
```

```python
#Tweet Query using tweet_id information from the df.Tweet IDs used to gather additional data(tweet_data) via Twitter's API
tweet_data = {}
count = 0
fails_dict = {}
start = timer()
for tweet in list(df.tweet_id):
    count += 1
    print(str(count) + ": " + str(tweet))
    try:
        tweet_status = api.get_status(tweet,
                                      wait_on_rate_limit=True, 
                                      wait_on_rate_limit_notify=True)
        tweet_data[str(tweet)] = tweet_status._json
    except: 
        print("Error for: " + str(tweet))
end = timer()
print(end - start)
```

```python
# save json data to text file
with open('tweet_json.txt', 'w') as outfile:  
    json.dump(tweet_data, outfile, 
              sort_keys = True,
              indent=4,
              ensure_ascii = True)
```

```python
# read json data from text file to data frame
tweet_df = pd.read_json('tweet_json.txt',orient='index')
tweet_df.head(2)
```

```python
df_clean.info()
```

```python
# Check if there is a floating rating from text column. We can see for example 13.5 for numerator
df[df.text.str.contains(r'(\d+(\.\d+))\/(\d+)')]['text'].values
```

```python
# Extract correct rating from text column
correct_rating = df_clean.text.str.extract('((?:\d+\.)?\d+)\/(\d+)', expand=True)
correct_rating
```

```python
correct_rating.columns= ['col1', 'col2']
correct_rating.head()
```

```python
correct_rating.info()
```

```python
#Replace the rating_numerator and rating_denominator by new correct values and convert them float
df_clean['rating_numerator']=correct_rating.col1.astype(float)
df_clean['rating_denominator']=correct_rating.col2.astype(float)
```

```python
# Test bot rating_numerator and rating_denominator are float
df_clean.info()
```

```python
# check whether those float numbers extracted from text were stored as foat correctly. for example firt text extraction in the array have 13.5 as its numerator.
df_clean[df_clean['rating_numerator'] == 13.5]
```

```python
#used to_datetime to convert the column from strings to DateTime objects 
df_clean['timestamp'] = pd.to_datetime(df_clean['timestamp'])
```

```python
#Test datetime type for timestamp
df_clean.info()
```

```python
#check nonretweet rows 
df_clean[df_clean.retweeted_status_id.isnull()]
```

```python
# select only some important columns and only nonretweet rows
cols=['tweet_id','timestamp','text','name','source','rating_numerator','rating_denominator','doggo','floofer','pupper','puppo']
df_clean=df_clean[df_clean.retweeted_status_id.isnull()][cols]
```

```python
#Test
df_clean.info()
```

```python
df_clean.columns
```

```python
# combine the last four columns into dog_stage column
df_clean['dog_stage'] = df_clean[df_clean.columns[-4:]].apply(lambda x: ','.join(x.dropna().astype(str)),
                                         axis=1)
df_clean['dog_stage'] = df_clean['dog_stage'].map(lambda x: x.lstrip('None,').rstrip(''))
df_clean.head()
```

```python
df_clean['dog_stage'].unique().tolist()
```

```python
map_stage={'': "None",
 'doggo,None,None,None' : "doggo",
 'puppo' : "puppo",
 'pupper,None':"pupper",
 'floofer,None,None': "pupper",
 'doggo,None,None,puppo': "pupper, puppo",
 'doggo,floofer,None,None': "doggo, floofer",
 'doggo,None,pupper,None': "floofer, pupper"}
```

```python
# replace values of dog_stage column based on map_stage
df_clean['dog_stage'].replace(map_stage, inplace=True)
```

```python
#test
df_clean.head()
```

```python
#test
df_clean['dog_stage'].unique()
```

```python
#Drop these columns for tidiness issue

df_clean.drop(['doggo','floofer','pupper','puppo'], axis=1, inplace=True)
```

```python
#test
df_clean.head()
```

```python
# clean up name column that are not true names and replace by "None" and capitalize names
map_name = {"a": "None", 'the': "None", 'an':'None'}
df_clean['name'].replace(map_name, inplace=True)
df_clean['name'] = df_clean['name'].str.title()
```

# clean the image_dataframe

```python
# Convert tweed_id column to string.

images_clean.tweet_id=images_clean.tweet_id.astype(str)
```

```python
images_clean.head()
```

```python
#Extract the true dog breed prediction from the three columns including algorithm confidence value. 
def predict_breed(row):
    if row.p1_dog:
        return row.p1
    elif row.p2_dog:
        return row.p2
    elif row.p3_dog:
        return row.p3
images_clean['dog_breed'] = images_clean.apply(lambda row: predict_breed(row),axis=1)
```

 # Inner merge  of the three dataframes on tweet_id column.

```python
#place the dataframes on a list. 
dfs = [df_clean, images_clean, tweet_df_clean]
#We can loop over the list,merging the dfs. But here reduce and a lambda is used.
result_df = reduce(lambda left,right: pd.merge(left,right,on='tweet_id'), dfs)

     
```
